```python
#Import dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Store filepaths into variable
CANNABIS_DATA = Path("Resources", "Canada_cannabis_data.csv")
```

```python
#Read in files.
cannabis_canada_df = pd.read_csv(CANNABIS_DATA)
```

```python
# Size of the original data
print(cannabis_canada_df.shape)
```

```python
cannabis_canada_df.head()
```

```python
# Removing columns which are not needed
Columns_delete =['unpackaged_production_units','unpackaged_inventory_units', 'unpackaged_production_kilograms', 'unpackaged_inventory_kilograms', 'packaged_production_kilograms',
                 'packaged_inventory_kilograms_-_federal_licence_holders', 'packaged_inventory_kilograms_-_provincial_distributors_and_reta', 
                 'sales_kilograms_-_medical', 'sales_kilograms_-_non-medical']
cannabis_refined_df = cannabis_canada_df.drop(Columns_delete, axis=1 )
cannabis_refined_df
```

```python
# Setting index to the _id column 
cannabis_refined_df.set_index('_id', inplace= True)
cannabis_refined_df
```

```python
#Removing name of index column
cannabis_refined_df.index.name= None
cannabis_refined_df
```

```python
#determining data type for each column
cannabis_refined_df.dtypes
```

```python
# convert year_-_month column to datetime using pd.to_datetime with predefined format
cannabis_refined_df['year_-_month'] = cannabis_refined_df['year_-_month'].str.strip()
cannabis_refined_df['year_-_month'] = pd.to_datetime(cannabis_refined_df['year_-_month'], format='%Y-%m')
cannabis_refined_df.head()
```

```python
#Retrieving list of columns
cannabis_refined_df.columns
```

```python
#Converting columns to numeric for calculations and graph plotting 
INTEGER_COLUMNS = ['packaged_production_units','packaged_inventory_units_-_federal_licence_holders',
       'packaged_inventory_units_-_provincial_distributors_and_retailer',
       'sales_units_-_medical', 'sales_units_-_non-medical' ]
cannabis_refined_df[INTEGER_COLUMNS] = cannabis_refined_df[INTEGER_COLUMNS].apply(pd.to_numeric, errors='coerce')
cannabis_refined_df
```

```python
# Extracting only the name in english before "/" in product_type column
cannabis_refined_df['product_type']= cannabis_refined_df['product_type'].str.split('/').str[0].str[:-1]
cannabis_refined_df
```

```python
#Renaming columns 
cannabis_refined_df= cannabis_refined_df.rename(columns={'year_-_month': "Date", 'product_type': "Product Type",
                                                         'packaged_production_units': "Packaged Production Units",
                                                         'packaged_inventory_units_-_federal_licence_holders': "Inventory- Federal License",
                                                         'packaged_inventory_units_-_provincial_distributors_and_retailer': "Inventory- Provincial",
                                                         'sales_units_-_medical': "Medical Sales Units", 'sales_units_-_non-medical': "Non Medical Sales Units"})
cannabis_refined_df.columns
```

Data Analysis 

```python
cannabis_refined_df
```

```python
product = cannabis_refined_df['Product Type'].unique()
print("The different Cannabis product are: " )
print(product)
```

```python
#Removing  year 2018 because of lack of data
cannabis_refined_df['Date'].dt.year
cannabis_refined_df = cannabis_refined_df.loc[cannabis_refined_df['Date'].dt.year > 2018]
cannabis_refined_df.drop(cannabis_refined_df.loc[cannabis_refined_df['Product Type'] =='whole cannabis plants' ].index, inplace=True)
cannabis_refined_df
```

```python
# Determining most used cannabis type based on the total sales
#creating a group by 
product_df = cannabis_refined_df .groupby("Product Type")
Total_product = product_df[['Packaged Production Units', 'Medical Sales Units','Non Medical Sales Units' ]].sum()
Total_product
```

```python
Most_used_medical= Total_product['Medical Sales Units'].max()
value= Total_product['Medical Sales Units'].idxmax()
value_recreational= Total_product['Non Medical Sales Units'].idxmax()
value_productio= Total_product['Packaged Production Units'].idxmax()

print("Type of cannabis most used for Medical purposes is : " + value)
print("Type of cannabis most used for Non Medical purposes is : " + value_recreational)
print("Type of cannabis with highest packaged production till date is : " + value_productio)
```

```python
#Creating a dataframe grouped by product type and year
product_year_df = cannabis_refined_df .groupby(["Product Type",cannabis_refined_df['Date'].dt.year])
Total_product_year = product_year_df[['Packaged Production Units', 'Medical Sales Units','Non Medical Sales Units' ]].sum()
Total_product_year
```

```python
#creating data frame for total production groupeb by product type and year
plot_production_year = product_year_df[['Packaged Production Units' ]].sum()
#creating a pivot table for plotting bargraph
new_df= plot_production_year.pivot_table(index= 'Product Type', columns= 'Date', values='Packaged Production Units')
new_df
```

```python

production_bar = new_df.plot(kind= 'bar', title='Production of different types of Cannabis over time ')
production_bar.set_xlabel('Product Types')
production_bar.set_ylabel("Packaged Production Units")
production_bar.ticklabel_format(axis='y', style='sci', scilimits=(0, 10000))
fig=production_bar.get_figure()
fig.set_size_inches(7,6)
fig.savefig("Images/total_production.png")



```

```python
##creating data frame for mediacl and non medical sales grouped by product type and year
Total_sales_year = product_year_df[['Medical Sales Units','Non Medical Sales Units' ]].sum()
Total_sales_year


```

```python
#Creating a pivot table for plotting a pie chart of medical sales of a product till date
Total_sales_medical = product_year_df[['Medical Sales Units']].sum()
medical_df= Total_sales_medical.pivot_table(index= 'Date', columns= 'Product Type', values='Medical Sales Units')
medical_df


```

```python
medical_df.columns
```

```python
#Creating pie chart for medical sales of product over different year
font_color = '#525252'
colors = ['#f7ecb0', '#ffb3e6', '#99ff99', '#66b3ff', '#c7b3fb','#ff6666', '#f9c3b7']
legend = ['cannabis extracts', 'cannabis seeds', 'cannabis topicals','dried cannabis', 'edible cannabis', 'vegetative cannabis plants']
fig, axes = plt.subplots(2, 3, figsize=(10, 10), facecolor='#e8f4f0')
fig.delaxes(ax= axes[1,2])
#Draw pie charts with legend
for i, (idx, row) in enumerate(medical_df.head(5).iterrows()):
    ax = axes[i // 3, i % 3]
    ax.pie(row, startangle=30, 
           wedgeprops=dict(width=.5), # For donuts
           colors=colors, 
           textprops={'color':font_color}, autopct='%1.00f%%')
    ax.set_title(idx, fontsize=16, color=font_color)
    
legend = plt.legend(legend, bbox_to_anchor=(1.3, 1), # Legend position
                    loc='upper left',  ncol=1, fancybox=True)
fig.subplots_adjust(wspace=.2) # Space between charts
title = fig.suptitle('Sales of different types of Cannabis for Medical purposes', y=.95, fontsize=20, color=font_color)
# To prevent the title from being cropped
plt.subplots_adjust(top=0.85, bottom=0.15)
plt.savefig("Images/sales_medical.png")
plt.show()
```

```python
#Creating a pivot table for plotting a pie chart of Non medical sales of the products till date
Total_sales_nonmedical = product_year_df[['Non Medical Sales Units']].sum()
nonmedical_df= Total_sales_nonmedical.pivot_table(index= 'Date', columns= 'Product Type', values='Non Medical Sales Units')
nonmedical_df
```

```python
#Creating pie chart for medical sales of product over different year
font_color = '#525252'
colors = ['#f7ecb0', '#ffb3e6', '#99ff99', '#66b3ff', '#c7b3fb','#ff6666', '#f9c3b7']
legend = ['cannabis extracts', 'cannabis seeds', 'cannabis topicals','dried cannabis', 'edible cannabis', 'vegetative cannabis plants']
fig, axes = plt.subplots(2, 3, figsize=(10, 10), facecolor='#e8f4f0')
fig.delaxes(ax= axes[1,2])
#Draw pie charts with legend
for i, (idx, row) in enumerate(nonmedical_df.head(6).iterrows()):
    ax = axes[i // 3, i % 3]
    ax.pie(row, startangle=30, 
           wedgeprops=dict(width=.5), # For donuts
           colors=colors, 
           textprops={'color':font_color}, autopct='%1.0f%%')
    ax.set_title(idx, fontsize=16, color=font_color)
legend = plt.legend(legend, bbox_to_anchor=(1.3, 0.87), # Legend position 
                    loc='upper left',  ncol=1, fancybox=True)
fig.subplots_adjust(wspace=.2) # Space between charts
title = fig.suptitle('Sales of different types of Cannabis for Non Medical Purposes', y=.95, fontsize=20, color=font_color)
# To prevent the title from being cropped
plt.subplots_adjust(top=0.85, bottom=0.15)
plt.savefig("Images/sales_nonmedical.png")
plt.show()
```
